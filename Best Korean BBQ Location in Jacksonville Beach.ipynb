### <span style="color:red"> I was unable to find any Korean BBQ restaurants anywhere in the general vicinity using Foursquare, so I decided to use Google maps to find the closest competitors and list them in a dataframe </span>

```python
competitors = pd.DataFrame()
competitors['Restaurant Name'] = ['Sam Won Garden Korean BBQ', "Kim's Korean BBQ Restaurant", 'Hot Pot & Korean BBQ', 'Korean BBQ House', 'Gangnam Korean Restaurant', 'Izakaya Pub']
competitors['Latitude'] = [30.278640, 30.199198, 30.285302, 30.189066, 30.296230, 30.321921]
competitors['Longitude'] = [-81.601786, -81.618078, -81.526291, -81.714023, -81.612586, -81.544219]
```

```python
competitors
```

```python
# create map of Jacksonville Beach using latitude and longitude values
map_jax = folium.Map(location=[latitude, longitude], zoom_start=11)

# add markers to map
for name, lat, lng, in zip(competitors['Restaurant Name'], competitors['Latitude'], competitors['Longitude']):
    label = folium.Popup(name, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_jax)
    
map_jax
```
